```python
import numpy as np

img_amps = [0.0, .46, .74, .92]
lightshifts = [0.,22.62e3, 34.19e3, 41.38e3]

import matplotlib.pyplot as plt

# Linear fit: lightshift = m * img_amp + b
m, b = np.polyfit(img_amps, lightshifts, 1)

# Plot data and fit
x_fit = np.linspace(min(img_amps), max(img_amps), 200)
y_fit = m * x_fit + b

plt.figure(figsize=(6, 4))
plt.scatter(img_amps, lightshifts, label="Data")
plt.plot(x_fit, y_fit, label=f"Linear fit: y = {m:.3e}x + {b:.3e}")
plt.xlabel("img_amp")
plt.ylabel("lightshift")
plt.legend()
plt.tight_layout()
plt.show()

def lightshift_from_img_amp(img_amp):
    return m * img_amp + b
```

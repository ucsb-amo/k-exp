```python
import numpy as np

# Extract x and y from the data points
data = [[0.,147.243],[.4,147.3],[.6,147.33],[.8,147.37]]
x_data = np.array([point[0] for point in data])
y_data = np.array([point[1] for point in data])

# Fit to a line (degree 1 polynomial)
coefficients = np.polyfit(x_data, y_data, 1)
m, b = coefficients  # slope and intercept

# Define the function
def y_from_x(x):
    return m * x + b

# Test it
print(f"Line equation: y = {m:.4f}*x + {b:.4f}")
print(f"Test: y(0.5) = {y_from_x(0.5):.4f}")
```

```python
import numpy as np

compression_factor = [.062,1.97,3.94]
detuning = [-517.5e6,-501.e6,-490.e6]

import matplotlib.pyplot as plt

# Linear fit: detuning = m * compression_factor + b
m, b = np.polyfit(compression_factor, detuning, 1)

x_data = np.array(compression_factor, dtype=float)
y_data = np.array(detuning, dtype=float)

x_fit = np.linspace(x_data.min(), x_data.max(), 200)
y_fit = m * x_fit + b

plt.figure(figsize=(6, 4))
plt.scatter(x_data, y_data, label="data", zorder=3)
plt.plot(x_fit, y_fit, label=f"fit: detuning = {m:.3e}*cf + {b:.3e}")
plt.xlabel("compression_factor")
plt.ylabel("detuning (Hz)")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

def detuning_from_compression_factor(cf):
    """Return fitted detuning (Hz) for a given compression factor."""
    cf = np.asarray(cf, dtype=float)
    return m * cf + b
```

```python
cfs = np.linspace(.16,3.94,10)
detunings = detuning_from_compression_factor(cfs)  # Example usage
```

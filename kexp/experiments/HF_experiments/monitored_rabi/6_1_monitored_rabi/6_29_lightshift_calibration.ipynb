```python
import numpy as np

img_amps = np.array([0.,.2,.3,.4,.5])
lightshifts = np.array([0.,40.e3,51.14e3,74.e3,94.e3])

import matplotlib.pyplot as plt

# Linear fit: lightshift = m * img_amp + b
m, b = np.polyfit(img_amps, lightshifts, 1)

# Plot data + fit
xfit = np.linspace(img_amps.min(), img_amps.max(), 200)
yfit = m * xfit + b

plt.figure(figsize=(5, 3.5))
plt.plot(img_amps, lightshifts, "o", label="data")
plt.plot(xfit, yfit, "-", label=f"fit: y = {m:.3e} x + {b:.3e}")
plt.xlabel("img_amp")
plt.ylabel("lightshift (Hz)")
plt.legend()
plt.tight_layout()
plt.show()

def lightshift_from_img_amp(img_amp):
    """Return fitted lightshift (Hz) for a given img_amp (scalar or array)."""
    return m * np.asarray(img_amp) + b
```

```python
imgamps = np.linspace(.1,.6,6)

119.4641e6 + lightshift_from_img_amp(imgamps)
```

```python
lightshift_from_img_amp(.4)
```

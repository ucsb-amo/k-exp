```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['font.family'] = 'Arial'

# Labels
title = "Rabi signal scaling"
xlabel = r"Pulse length $(2\pi/\Omega)$"
ylabel = r"$S_z$"

t = np.linspace(0, 1, 500)
y = (t - 1)**2 - t**2
y1 = (1 - t)**2
y2 = -t**2

fig, ax = plt.subplots(figsize=(5, 4))
ax.plot(t, y,  label=r"$N_{\uparrow} - N_{\downarrow}$")
ax.plot(t, y1, label=r"$N_{\uparrow}$")
ax.plot(t, y2, label=r"$N_{\downarrow}$")
ax.set_title(title)
ax.set_xlabel(xlabel)
ax.set_ylabel(ylabel)
ax.axhline(0, color='black', linewidth=0.5, linestyle='--')
ax.legend()
# ax.grid()
plt.tight_layout()
plt.show()
```

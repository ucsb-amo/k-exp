# Raman Test Data Analysis

Load and plot the v data saved from the raman_test_continuous_mode experiment, showing differences in v values versus xvar.

## Import Required Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Set plotting parameters
plt.rcParams['figure.dpi'] = 150
plt.rcParams['figure.figsize'] = (10, 6)
```

## Load Data from Saved Files

```python
# Load the saved data files
# You can modify these file paths and run_id as needed
run_id = 0  # Change this to load data from a different run

xvars_file = f'./data_{run_id}_xvars.npy'
v_file = f'./data_{run_id}_v.npy'

# Load the arrays
xvars = np.load(xvars_file)
v_data = np.load(v_file)

print(f"Loaded data from run {run_id}")
print(f"xvars shape: {xvars.shape}")
print(f"v_data shape: {v_data.shape}")
print(f"\nxvars (first 10): {xvars[:10]}")
print(f"v_data (first 5): {v_data[:5]}")
```

## Calculate Differences in v Values

```python
# Calculate differences in v values
# Option 1: Difference from the first value
v_diff_from_first = v_data - v_data[0]

# Option 2: Consecutive differences (derivative-like)
v_diff_consecutive = np.diff(v_data)

# Option 3: Difference from the mean
v_mean = np.mean(v_data)
v_diff_from_mean = v_data - v_mean

print(f"Mean v value: {v_mean:.6e}")
print(f"Max difference from first: {np.max(np.abs(v_diff_from_first)):.6e}")
print(f"Max consecutive difference: {np.max(np.abs(v_diff_consecutive)):.6e}")
print(f"Std of v values: {np.std(v_data):.6e}")
```

## Plot v Differences vs xvar

```python
# Create a comprehensive plot showing v differences vs xvar
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: v values vs xvar
ax = axes[0, 0]
ax.plot(xvars, v_data, 'o-', linewidth=2, markersize=6)
ax.set_xlabel('xvar values')
ax.set_ylabel('v values')
ax.set_title('v Data vs xvar')
ax.grid(True, alpha=0.3)

# Plot 2: Difference from first value vs xvar
ax = axes[0, 1]
ax.plot(xvars, v_diff_from_first, 'o-', linewidth=2, markersize=6, color='tab:orange')
ax.set_xlabel('xvar values')
ax.set_ylabel('Δv (from first)')
ax.set_title('v Difference from First Value vs xvar')
ax.grid(True, alpha=0.3)
ax.axhline(y=0, color='k', linestyle='--', alpha=0.3)

# Plot 3: Consecutive differences vs xvar (offset by one)
ax = axes[1, 0]
ax.plot(xvars[:-1], v_diff_consecutive, 'o-', linewidth=2, markersize=6, color='tab:green')
ax.set_xlabel('xvar values')
ax.set_ylabel('Δv (consecutive)')
ax.set_title('Consecutive v Differences vs xvar')
ax.grid(True, alpha=0.3)
ax.axhline(y=0, color='k', linestyle='--', alpha=0.3)

# Plot 4: Difference from mean vs xvar
ax = axes[1, 1]
ax.plot(xvars, v_diff_from_mean, 'o-', linewidth=2, markersize=6, color='tab:red')
ax.set_xlabel('xvar values')
ax.set_ylabel('Δv (from mean)')
ax.set_title('v Difference from Mean vs xvar')
ax.grid(True, alpha=0.3)
ax.axhline(y=0, color='k', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nPlot created for run {run_id}")
```
